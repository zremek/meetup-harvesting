```python
import meetup.api
import json
import requests
import time
import codecs
import sys
import io
```

Warning: *this code does not work!*


To make it work you need a meetup.com API token. 
Meetup members can easily get their token at https://secure.meetup.com/meetup_api/key/

Just swap "MY_API_KEY" in a request.get("key=") to the proper token, and you're ready to go.

Data that I got represents meetups in Poland, topic "data science", 2019-06-04.

```python
ds_groups_29 = requests.get(
    'https://api.meetup.com/find/groups?key=MY_API_KEY&photo-host=public&country=PL&topic_id=102811&radius=100.0&page=200'
                       )
ds_groups_29.json()
# gets 29 groups, while total 86 in Poland https://www.meetup.com/topics/data-science/pl/
# reason is my location - Lodz
```

```python
data = ds_groups_29.json()
try:
 to_unicode = unicode
except NameError:
 to_unicode = str
with io.open('ds_groups_29.json', 'w', encoding='utf8') as outfile:
 str_ = json.dumps(data,
 indent=4, sort_keys=True,
 separators=(',', ': '), ensure_ascii=False)
 outfile.write(to_unicode(str_))
```
